```python
#We import the library Matplotlib to access the Charts
from matplotlib import pyplot as plt

#Now we import the library Pandas to access the CSV data
import pandas as pd

#My research is on how many military equipment russia lost in Ukraine war from February 25,2022 to May 08, 2022

#We name our CSV data as mydata, I used the github method and copy pasted the URL to access the data
mydata = pd.read_csv('https://raw.githubusercontent.com/ilyasrasooliar/assignment/main/russia_losses_equipment.csv')
#print(mydata)

#I typed the code mydata.columns to see how many columns we have an it makes it easy to access the columns
mydata.columns

#Showing the Equipments numbers in Scatter chart for each equipment loss

#Since our columns are many, we need to change the chart/plt size
plt.figure(figsize=(24,12))

#We are accessing each columns individually
plt.scatter(mydata['day'], mydata['special equipment'], label='Special Equipment Numbers')
plt.scatter(mydata['day'], mydata['mobile SRBM system'], label='Mobile SRBM System Numbers')
plt.scatter(mydata['day'], mydata['vehicles and fuel tanks'], label='Vehicles and Fuel Tanks Numbers')
plt.scatter(mydata['day'], mydata['cruise missiles'], label='Cruise Missiles Numbers')
plt.scatter(mydata['day'], mydata['aircraft'], label='Aircraft Numbers')
plt.scatter(mydata['day'], mydata['helicopter'], label='Helicopter Numbers')
plt.scatter(mydata['day'], mydata['tank'], label='Tanks Numbers')
plt.scatter(mydata['day'], mydata['APC'], label='APC Numbers')
plt.scatter(mydata['day'], mydata['field artillery'], label='Field Artillery Numbers')
plt.scatter(mydata['day'], mydata['MRL'], label='MRL Numbers')
plt.scatter(mydata['day'], mydata['military auto'], label='Military Auto Numbers')
plt.scatter(mydata['day'], mydata['fuel tank'], label='Fuel tank Numbers')
plt.scatter(mydata['day'], mydata['drone'], label='Drone Numbers')
plt.scatter(mydata['day'], mydata['naval ship'], label='Naval Ship Numbers')
plt.scatter(mydata['day'], mydata['anti-aircraft warfare'], label='Anti-aircraft Warfare Numbers')

#To create lables we need to put the plt.legend()
plt.legend()
plt.show()
```

```python
#Now we create a Pie Chart

df = pd.read_csv('https://raw.githubusercontent.com/ilyasrasooliar/assignment/main/russia_losses_equipment.csv')

#To create a Pie chart out of our data and compare the percentages of 
#equipments lost we need to sum all the numbers of euipments column-wise
aircraft = df['aircraft'].sum()
helicopter = df['helicopter'].sum()
tank = df['tank'].sum()
apc = df['APC'].sum()
field_artillery = df['field artillery'].sum()
MRLk = df['MRL'].sum()
military_auto = df['military auto'].sum()
fuel_tank = df['fuel tank'].sum()
drone = df['drone'].sum()
naval_ship = df['naval ship'].sum()
anti_aircraft_warfare = df['anti-aircraft warfare'].sum()
special_equipment = df['special equipment'].sum()
mobile_SRBM_system = df['mobile SRBM system'].sum()
vehicles_and_fuel_tanks = df['vehicles and fuel tanks'].sum()
cruise_missiles = df['cruise missiles'].sum()

mypiec = [aircraft, helicopter, tank, apc, field_artillery, MRLk, military_auto, fuel_tank, drone, naval_ship,
          anti_aircraft_warfare, special_equipment, mobile_SRBM_system, vehicles_and_fuel_tanks, cruise_missiles] #here we are putting our data set
lost_equipments = ('aircraft', 'helicopter', 'tank', 'apc', 'field_artillery', 'MRLk', 'military_auto', 'fuel_tank', 'drone', 
                   'naval_ship', 'anti_aircraft_warfare','special_equipment', 'mobile_SRBM_system', 
                   'vehicles_and_fuel_tanks', 'cruise_missiles')
dist = (0,0,0,0.1,0,0,0,0.8,0,0.8,0.15,0.8,0,0,1)
#print(df.columns) #This prints the columns
#print(df) #This prints our table

print(aircraft) #we can see that the total of aircraft Russia lost is 9,058
print(fuel_tank) #we can see that the total of fuel tanks Russia lost is 4,506

#Now we are putting our data into a pie chart 
fig1, ax1 = plt.subplots(figsize=(40,28))
ax1.pie(mypiec, autopct='%1.1f%%' ,textprops={'fontsize': 16}, startangle=45,
        labels= lost_equipments, rotatelabels=True, pctdistance=0.6, shadow=False, explode=dist)
plt.title('Number of Equipments Russian Lost in Ukraine War')
ax1.axis()
plt.show()
```
