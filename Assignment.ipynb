```python
#We import the library Matplotlib to access the Charts
from matplotlib import pyplot as plt

#Now we import the library Pandas to access the CSV data
import pandas as pd
```

```python
#My research is on how many military equipment russia lost in Ukraine war from February 25,2022 to May 08, 2022
#We name our CSV data as mydata, I used the github method and copy pasted the URL to access the data

mydata = pd.read_csv('https://raw.githubusercontent.com/ilyasrasooliar/assignment/main/russia_losses_equipment.csv')
print(mydata)

#I typed the code mydata.columns to see how many columns we have an it makes it easy to access the columns
mydata.columns
```

```python
plt.figure(figsize=(12,8))
plt.scatter(mydata['day'], mydata['special equipment'], label='Special Equipment Numbers')
plt.scatter(mydata['day'], mydata['mobile SRBM system'], label='Mobile SRBM System Numbers')
plt.scatter(mydata['day'], mydata['vehicles and fuel tanks'], label='Vehicles and Fuel Tanks Numbers')
plt.scatter(mydata['day'], mydata['cruise missiles'], label='Cruise Missiles Numbers')
plt.scatter(mydata['day'], mydata['aircraft'], label='Aircraft Numbers')
plt.scatter(mydata['day'], mydata['helicopter'], label='Helicopter Numbers')
plt.scatter(mydata['day'], mydata['tank'], label='Tanks Numbers')
plt.scatter(mydata['day'], mydata['APC'], label='APC Numbers')
plt.scatter(mydata['day'], mydata['field artillery'], label='Field Artillery Numbers')
plt.scatter(mydata['day'], mydata['MRL'], label='MRL Numbers')
plt.scatter(mydata['day'], mydata['military auto'], label='Military Auto Numbers')
plt.scatter(mydata['day'], mydata['fuel tank'], label='Fuel tank Numbers')
plt.scatter(mydata['day'], mydata['drone'], label='Drone Numbers')
plt.scatter(mydata['day'], mydata['naval ship'], label='Naval Ship Numbers')
plt.scatter(mydata['day'], mydata['anti-aircraft warfare'], label='Anti-aircraft Warfare Numbers')
plt.legend()
plt.show()
```
